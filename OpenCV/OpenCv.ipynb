Drawing on images

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(384,10),pt2=(500,150),color=(0,255,0), thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,255,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,100),radius=50,color=(255,0,0),thickness=8)
```

```python
plt.imshow(blank_img)
```

```python
from numpy import blackman

cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=4)
```

```python
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
```

```python
plt.imshow(blank_img)
```

```python
import cv2
import numpy as np

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)

cv2.namedWindow('my_drawing')

cv2.setMouseCallback('my_drawing',draw_circle)

img = np.zeros((512,512,3))


while True:
    cv2.imshow('my_drawing',img)

    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()
```
